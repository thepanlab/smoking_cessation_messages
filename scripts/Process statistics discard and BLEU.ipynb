# vB: Prompt selection

## BLEU-4

We would use BLEU-4 with threshold 0.5


```python
df_prompt_bleu4 = pd.read_csv("/home/pcallec/smoking_cessation_messages/results/vB/v2/self-BLEU-4/reduced/statistics_selfBLEU4.csv",
                              index_col = 0)
```

```python
df_prompt_bleu4.head()
```

```python
df_prompt_bleu4["filename"].unique()
```

```python
def get_prompt_version(str_type):

    items = str_type.split('_')
    
    prompt_version = items[2]
        
    return prompt_version

def get_model(str_type):
    items = str_type.split('_')
    
    model = items[5]
        
    return model
```

```python
df_prompt_bleu4.insert(1, "model", df_prompt_bleu4["filename"].apply(get_model))
df_prompt_bleu4.insert(2, 'prompt_version', df_prompt_bleu4["filename"].apply(get_prompt_version))
```

```python
df_prompt_bleu4.head()
```

```python
df_prompt_bleu4 = df_prompt_bleu4.drop(columns=["filename"])
df_prompt_bleu4.head()
```

```python
df_prompt_bleu4_th05 = df_prompt_bleu4.query("threshold == 0.50")
df_prompt_bleu4_th05 = df_prompt_bleu4_th05.drop(columns=["threshold"])
```

```python
df_prompt_bleu4_th05.head()
```

```python
dict_new_names = {"n_remaining":"BLEU4_remaining", "percentage_discarded":r"BLEU4_%remaining"}
df_prompt_bleu4_th05 = df_prompt_bleu4_th05.rename(columns=dict_new_names)
```

```python
df_prompt_bleu4_th05.head()
```

```python
df_prompt_bleu4_th05[r"BLEU4_%discarded"] = 1 - df_prompt_bleu4_th05[r"BLEU4_%remaining"]
```

```python
df_prompt_bleu4_th05.head()
```

```python
df_prompt_bleu4_th05["model"] = df_prompt_bleu4_th05["model"].apply(str.upper)
```

```python
df_prompt_bleu4_th05.head()
```

```python
df_prompt_bleu4_th05.insert(0, "model_prompt_version", df_prompt_bleu4_th05['model'] +"_"+ df_prompt_bleu4_th05["prompt_version"])
```

```python
df_prompt_bleu4_th05.head()
```

```python
df_prompt_bleu4_th05.dtypes
```

# temp

```python
df_stats = pd.read_csv('/home/pcallec/smoking_cessation_messages/results/vB/v2/statistics_discard_and_bleu.csv',
            header = [0], index_col = 0) 

df_stats.head()
```

```python
df_stats.columns
```

## means

```python
df_stats_mean = df_stats.groupby('prompt version').mean()
df_stats_mean
```

```python
l_columns = df_stats_mean.columns
```

```python
l_columns[0]
```

```python
l_new_column_names = []
d_change_col_names = dict()
for i_column in l_columns:
    new_column_name = i_column +  "-mean"
    l_new_column_names.append(new_column_name)
    d_change_col_names[i_column] = new_column_name
```

```python
l_new_column_names
```

```python
d_change_col_names
```

```python
df_stats_mean.rename(columns = d_change_col_names, inplace=True)
```

```python
df_stats_mean.columns
```

```python
df_stats_mean
```

## std error

```python
df_stats_std = df_stats.groupby('prompt version').std(ddof = 1)
df_stats_std
```

```python
# 5 because of 5 models: gptj6b, bloom, opt6.7b, opt13b, opt30b
df_stats_std_err = df_stats_std/5**0.5
```

```python
df_stats_std_err
```

```python
l_columns = df_stats_std_err.columns
```

```python
l_new_column_names = []
d_change_col_names = dict()
for i_column in l_columns:
    new_column_name = i_column +  "-err"
    l_new_column_names.append(new_column_name)
    d_change_col_names[i_column] = new_column_name
```

```python
df_stats_std_err.rename(columns = d_change_col_names, inplace=True)
```

```python
df_stats_std_err
```

```python
df_stats_mean_std_err = df_stats_mean.join(df_stats_std_err)
df_stats_mean_std_err
```

```python
l_col_mean_std_err = df_stats_mean_std_err.columns
l_col_mean_std_err_ordered = []

n_real_columns = int(len(l_col_mean_std_err)/2)

for i in range(n_real_columns):
    l_col_mean_std_err_ordered.extend([l_col_mean_std_err[i],l_col_mean_std_err[n_real_columns+i]])
```

```python
df_stats_mean_std_err = df_stats_mean_std_err[l_col_mean_std_err_ordered]
```

```python
df_stats_mean_std_err.to_csv("/home/pcallec/smoking_cessation_messages/results/vB/v2/stats_filters/stats_filters_vB_v2.csv")
```
