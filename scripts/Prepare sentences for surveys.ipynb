```python
def get_model(str_type):
    model = None
    decoding_version = None
    
    if str_type != "original":
        items = str_type.split('_')
        
        model = items[0]
        decoding_version = items[1]
    else:
        model = "human"
    return model

def get_decoding_version(str_type):
    model = None
    decoding_version = None
    
    if str_type != "original":
        items = str_type.split('_')
        
        model = items[0]
        decoding_version = items[1]
        
    else:
        decoding_version = "human"
        
    return decoding_version
```
