```python
def pandas_fill_empty(df_to_fill, n_rows, index_column_name,
                      property_column_name):
    # Create empty dataframe

    data = {index_column_name : np.arange( n_rows), 
            property_column_name: [0]* n_rows}
    df_temp_blank = pd.DataFrame(data)
    df_temp_blank = df_temp_blank.set_index(index_column_name)
    
    
    df_to_fill_filled = df_to_fill.merge(df_temp_blank, left_on=index_column_name, right_on=index_column_name,
                        suffixes=('', '_right'), how="outer")   
    df_to_fill_filled = df_to_fill_filled.drop(columns = [f"{property_column_name}_right"])
    df_to_fill_filled = df_to_fill_filled.sort_index()
    
    df_to_fill_filled = df_to_fill_filled.fillna(value={index_column_name: 0})

    return df_to_fill_filled
```
