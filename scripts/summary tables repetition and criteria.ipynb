```python
l_columns = ["prompt_version", "model",
                "Message", "index_prompt", "message_position"]

df_all_messages = pd.DataFrame(columns=l_columns)

for pathfile in l_generated_files:
    path_temp = Path(pathfile)
    name_temp = path_temp.name
    prompt_version = name_temp.split("_")[2]
    model = name_temp.split("_")[5]
    
    df_temp = pd.read_csv(pathfile, index_col=0)
      
    df_temp.insert(0, "prompt_version", prompt_version)
    df_temp.insert(1, "model", model)   
    
    df_all_messages = pd.concat([df_all_messages, df_temp], ignore_index=True)
```

```python
df_messages_count = df_all_messages.groupby(["prompt_version", "model", "index_prompt"]).count()
df_messages_count.head()
```

```python
df_messages_count.reset_index(inplace=True, drop=False)
```

```python
df_messages_count_subset = df_messages_count[["prompt_version", "model",
                                              "index_prompt", "Message"]]
```

```python
df_messages_count_subset.tail()
```

```python
# Create empy dataframe
# Create empy dataframe
N_PROMPTS = 5
N_MODELS = 5
N_INDICES = 100
INITAL_N_MESSAGES = 0

l_versions = ["v1","v2","v3","v4","v5"]
l_versions_df = []

for version in l_versions:
   l_versions_df.extend([version]*N_MODELS*N_INDICES)

l_models = ['gpt-j-6B', 'bloom-7b1', 'opt-6.7b', 'opt-13b', 'opt-30b']

l_models_df = []
for i in range(N_PROMPTS):
   l_model_temp = []
   for model in l_models:
      l_model_temp.extend([model]*N_INDICES)
      
   l_models_df.extend(l_model_temp)

l_indices_df = []
for version in l_versions:
   for model in l_models:
      l_indices_df.extend(np.arange(N_INDICES))

l_n_messages = [0]*N_PROMPTS*N_MODELS*N_INDICES

data = {'prompt_version' : l_versions_df, 
        'model': l_models_df,
        'index_prompt': l_indices_df,
        "n_messages": l_n_messages}
df_prompt_blank = pd.DataFrame(data)
# df.index = df["index_prompt"]
df_prompt_blank = df_prompt_blank.set_index("index_prompt")
```

```python
df_prompt_blank.head()
```

```python
df_messages_count_subset_filled = df_messages_count_subset.merge(df_prompt_blank, left_on=["prompt_version","model","index_prompt"],
                                                                right_on=["prompt_version","model","index_prompt"],
                                                                suffixes=('', '_right'), how="outer")
df_messages_count_subset_filled
```

```python
df_messages_count_subset_filled = df_messages_count_subset_filled.drop(columns = "n_messages_right")
df_messages_count_subset_filled.head()
```

```python
df_messages_count_subset_filled = df_messages_count_subset_filled.fillna(value={"n_messages": 0})
df_messages_count_subset_filled.head()
```

```python
df_prompt_mean = df_messages_count_subset_filled.groupby("prompt_version").mean()
df_prompt_mean
```

```python
df_prompt_mean
```

```python
df_prompt_mean = df_prompt_mean.drop(columns = "index_prompt")
df_prompt_mean = df_prompt_mean.rename(columns = {"n_messages":"n_messages_mean"})
df_prompt_mean
```

```python
df_prompt_std = df_messages_count_subset_filled.groupby("prompt_version").std()
df_prompt_std
```

```python
df_prompt_std = df_prompt_std.drop(columns = "index_prompt")
df_prompt_std = df_prompt_std.rename(columns = {"n_messages":"n_messages_std"})
df_prompt_std
```

```python
df_prompt_std_err = df_messages_count_subset_filled.groupby("prompt_version").std()/(N_MODELS*N_INDICES)**0.5
df_prompt_std_err
```

```python
df_prompt_std_err = df_prompt_std_err.drop(columns = "index_prompt")
df_prompt_std_err = df_prompt_std_err.rename(columns = {"n_messages":"n_messages_std_err"})
df_prompt_std_err
```

```python
df_prompt_all = pd.concat([df_prompt_mean, df_prompt_std, df_prompt_std_err], axis = 1)
df_prompt_all
```

```python
df_prompt_all["mean_over_std"] = df_prompt_all.eval("n_messages_mean/n_messages_std")
```
