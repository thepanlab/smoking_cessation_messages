```python
l_feature = ["style", "accuracy", "credibility", "persuasion", "revision_option", "revision",
            "timing_First Click", "timing_Last Click", "timing_Page Submit", "timing_Click Count"]
```

```python
l_columns = [[f"M{i}_{feature}" for feature in l_feature] for i in range(1,51) ]
```

```python
l_file_path = ["/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q1_May 19, 2023_14.04.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q2_May 19, 2023_14.08.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q3_May 26, 2023_16.46.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q4_May 26, 2023_16.47.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q5_May 26, 2023_16.49.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q6_May 26, 2023_16.49.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q7_May 26, 2023_16.51.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q8_May 26, 2023_16.51.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q9_May 26, 2023_16.52.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q10_May 26, 2023_16.53.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q11_May 26, 2023_16.55.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q12_May 26, 2023_16.55.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q13_May 26, 2023_16.56.csv",
               "/home/pcallec/smoking_cessation_messages/data/qualtrics/Human Q14_June 12, 2023_11.20.csv"]
l_index = [3,3,4,3,3,
           4,4,4,3,3,
           3,3,3,3]
l_output = ["/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q1_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q2_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q3_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q4_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q5_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q6_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q7_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q8_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q9_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q10_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q11_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q12_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q13_results.csv",
            "/home/pcallec/smoking_cessation_messages/results/reformat_results_qualtrics/q14_results.csv"]

for i in range(len(l_file_path)):
    df_q = pd.read_csv(l_file_path[i])
    
    df_q_subset = df_q.iloc[[l_index[i]],:]
    df_q_subset = df_q_subset.reset_index(drop=True)

    # Splitting data in rows

    df_q_new = pd.DataFrame()

    for index_question in range(50):

        # create column index_question
        # change column name
        df_temp =  df_q_subset.loc[[0],l_columns[index_question]]
        
        # Create dict iteratively
        dicts = {}
        keys = [f"M{index_question+1}_{feature}" for feature in l_feature]

        for j in range(len(l_feature)):
            dicts[keys[j]] = l_feature[j]
            
        # break 

        df_temp = df_temp.rename(columns=dicts)
        
        df_temp["question"] = index_question + 1 
        
        l_new_columns = ["question"] + l_feature
        df_temp = df_temp[l_new_columns]
        # break
        
        df_q_new = pd.concat([df_q_new, df_temp ])
        
    df_q_new = df_q_new.reset_index(drop=True)
    # df_q_new.head()

    df_q_new .to_csv(l_output[i])
```
